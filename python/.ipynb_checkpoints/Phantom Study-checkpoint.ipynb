```python
def estimate_width(results):
    nerve_plot = np.array([[np.median(r[2].nerve_width[:,1]), np.median(np.abs(r[2].edge_values.flatten()))] for r in results if r[2] is not None and r[2].nerve_width is not None])
    lower = round(nerve_plot.shape[0] * 0.02)
    upper = round(nerve_plot.shape[0] * 0.1)
    sort_idx = np.flip(nerve_plot[:,1].argsort())
    good_pts = nerve_plot[sort_idx[lower:upper], :]
#     plt.scatter(good_pts[:,0], good_pts[:,1])
    estimate = np.mean(good_pts[:,0])
    return estimate
```

```python
import re
from glob import glob
import pickle
import pandas as pd
import numpy as np


pattern = r'(?P<size_id>\d+)mm-capture_(?P<sample_id>\d+)-out'
files = glob('*/width.p')
tmp = []
for fp in files:
    size_id, sample_id = re.search(pattern, fp).group('size_id', 'sample_id')
    with open(fp, 'rb') as f:
        width = pickle.load(f)
    tmp.append([int(size_id), sample_id, width])
df = pd.DataFrame(tmp, columns=['size_id', 'sample_id', 'width'])
    
```

```python
import re
from glob import glob
import pickle
import pandas as pd
import numpy as np
import os
from ocularus import *

pattern = r'(?P<size_id>\d+)mm-capture_(?P<sample_id>\d+)-out'
dirs = os.listdir('.')
result_fps = []
for fp in dirs:
#     size_id, sample_id 
    found = re.search(pattern, fp)
    if found is not None:
        size_id, sample_id = found.group('size_id', 'sample_id')
        result_fps.append((fp, size_id, sample_id))
        
```

```python
best_frames = dict()

def load_best_frames(fp):
    fp2 = fp + '/best_frames.p'
    if os.path.exists(fp2):
        with open(fp2, 'rb') as f:
            return pickle.load(f)
    else:
        return dict()

def write_best_frames(best_frames, fp):
    fp2 = fp + '/best_frames.p'
    with open(fp2, 'wb') as f:
        pickle.dump(best_frames, f)
```

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets
from IPython.display import display
# "global" variables into interact

def load_new(fp):
#     print(fp)
    results = load_results(fp)
#     print(results)
    frame_idx = 0
    best_frames = load_best_frames(fp)
    return [fp, results, frame_idx, best_frames]

current_data = load_new(result_fps[0][0])

def on_button_clicked(b):
    global current_data
    fp = current_data[0]
    frame_idx = current_data[2]
    best_frames = current_data[3]
    if frame_idx in best_frames:
        best_frames.pop(frame_idx)
    else:
        best_frames[frame_idx] = dict()
    write_best_frames(best_frames, fp)

button = ipywidgets.Button(description="Toggle Frame")
button.on_click(on_button_clicked)
output = ipywidgets.Output()
    
def foo(result_fp, frame_idx):
    global current_data, button, output
    fp = result_fp[0]
    if fp != current_data[0]:
        current_data = load_new(fp)
    
    current_data[2] = frame_idx
    results = current_data[1]
#     print(results)
    if frame_idx < len(results):
        img = results[frame_idx][0]
        plt.imshow(img)
        display(button, output)
        if frame_idx in current_data[3]:
            print("Best frame")
plt.rcParams['figure.figsize'] = [10, 10]
interact(foo, result_fp=result_fps, frame_idx=(0, 500))
```

```python
tmp = []
for rfp in result_fps:
    print(fp)
    fp, size_id, sample_id = rfp
    data = load_new(fp)
    results = data[1]
    best_results = [results[i] for i in data[3].keys()]
    try:
        best_width = estimate_width(best_results, lower_perc=0, upper_perc=1)
        tmp.append((size_id, sample_id, best_width))
        
    except:
        pass
```

```python
df3 = pd.DataFrame(tmp, columns=['size_id', 'sample_id', 'best_width'])
df3['size_id'] = df3['size_id'].astype('float32')
plt.scatter(df3['size_id'], df3['best_width'])
```
